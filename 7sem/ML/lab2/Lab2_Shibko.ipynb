```python
sns.set_style('darkgrid')
sns.countplot(y='Year',data=data,palette='colorblind')
plt.xlabel('Number')
plt.ylabel('Year')
plt.show()
```

```python
data['Year'] = data['Year'].astype(str)  # Преобразование в строку
count2001 = data[data['Year'] == '2001'].shape[0]
countDV = len(data[data['DV'] == 1])  # Предполагается, что DV обозначается как 1 для случаев домашнего насилия
print("\n Total number of crimes in 2001:", count2001)
print("\n Total number of DV:", countDV)
```

```python
#
sns.set_style('whitegrid')
sns.countplot(x='Year',data=data,palette='colorblind')
plt.xlabel("Year")
plt.ylabel("Number")
plt.show()
```

```python
sns.set_style('darkgrid')
plt.title('Relationship between Rapes and DD',size=16)
plt.xlabel('Rape',size=12)
plt.ylabel('DD',size=12)
sns.scatterplot(x='Rape', y='DD', data=data, hue='Year', edgecolor='black', palette='cubehelix')
plt.show()
```

```python
size_data = data[['Rape', 'AoW', 'DD']]

sns.set_style("darkgrid")
sns.kdeplot(data=size_data,shade=True, palette='colorblind')
plt.show()
```
